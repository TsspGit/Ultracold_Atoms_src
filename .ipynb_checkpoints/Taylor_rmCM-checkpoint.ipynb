```python
__author__ = "@Tssp"
__date__ = "6/10/20"
import sympy as sp
from sympy import init_session
```

```python
x1, x2, mu1, mu2, kx, Vox1, Vox2, x, X = sp.symbols('x_1 x_2 \mu_1 \mu_2 k_x V^1_o_x V^2_o_x x X')
```

```python
VoL = Vox1*sp.sin(kx*x1)**2 + Vox2*sp.sin(kx*x2)**2
VoL_expanded = sp.series(VoL, x1, n=8).removeO()
VoL_expanded = sp.series(VoL_expanded, x2, n=8).removeO()
VoL_expanded
```

```python
VoL_rmCM = VoL_expanded.subs(x2, X - x*mu1)
VoL_rmCM = VoL_rmCM.subs(x1, X + x*mu2)
VoL_rmCM
```

```python
VoL_rmCM_expanded = sp.expand(VoL_rmCM)
VoL_rmCM_expanded
```

```python
VoL_rmCM_expanded.coeff(x, 2)
```
