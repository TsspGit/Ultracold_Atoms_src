```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt
```

```python
phi = np.linspace(0, 2*pi, num=100)
theta = np.linspace(0, pi, num=100)
r = np.linspace(0, 1.5, num=100)
[R, Phi] = np.meshgrid(r, phi)
```

```python
def paraboloid(R, Phi):
    X = R*np.cos(Phi)
    Y = R*np.sin(Phi)
    Z = R**2
    return X, Y, Z

def sphere(r, phi, theta):
    [Phi, Theta] = np.meshgrid(phi, theta)
    X = r*np.cos(Phi) * np.sin(Theta)
    Y = r*np.sin(Phi) * np.sin(Theta)
    Z = r*np.cos(Theta)
    return X, Y, Z
```

# 3D-3D

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
ax.plot_surface(paraboloid(R, Phi)[0], paraboloid(R, Phi)[1], paraboloid(R, Phi)[2], alpha=0.4, color='orangered')
ax.plot_surface(paraboloid(R, Phi)[0], paraboloid(R, Phi)[1], paraboloid(R, Phi)[2], alpha=0.4)
ax.plot_surface(sphere(0.2, phi, theta)[0], sphere(0.2, phi, theta)[1], sphere(0.2, phi, theta)[2] + 0.25, color='k')
ax.view_init(elev=20, azim=0)
ax.set_xticks(())
ax.set_yticks(())
ax.axis('off')
plt.savefig('3D-3D_trap.png', dpi=200)
```

# 3D-1D

```python
def cilinder(phi):
    r = np.linspace(-1.5, 1.5, num=100)
    [R, Phi] = np.meshgrid(r, phi)
    X = R
    Y = 0.5*np.cos(Phi)
    Z = 0.5*np.sin(Phi)
    return X, Y, Z
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
ax.plot_surface(sphere(0.2, phi, theta)[0], sphere(0.2, phi, theta)[1], sphere(0.2, phi, theta)[2] + 0.25, color='k')
ax.plot_surface(cilinder(phi)[0], cilinder(phi)[1], cilinder(phi)[2]+0.25, alpha=0.4)
ax.plot_surface(paraboloid(R, Phi)[0], paraboloid(R, Phi)[1], paraboloid(R, Phi)[2], alpha=0.4, color='orangered')
ax.view_init(elev=20, azim=40)
ax.set_xticks(())
ax.set_yticks(())
ax.axis('off')
plt.savefig('3D-1D_trap.png', dpi=200)
```

# 1D-1D

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
ax.plot_surface(sphere(0.2, phi, theta)[0], sphere(0.2, phi, theta)[1], sphere(0.2, phi, theta)[2], color='k')
ax.plot_surface(10*cilinder(phi)[0], cilinder(phi)[1], cilinder(phi)[2], alpha=0.4, color='orangered')
ax.plot_surface(10*cilinder(phi)[0], cilinder(phi)[1], cilinder(phi)[2], alpha=0.4)
ax.view_init(elev=20, azim=40)
ax.set_zlim(-1, 1)
ax.set_ylim(-1.5, 1.5)
ax.set_xticks(())
ax.set_yticks(())
ax.axis('off')
plt.savefig('1D-1D_trap.png', dpi=200)
```
