```python
__author__ = "@Tssp"
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
plt.rc('text',usetex=True)
plt.rc('font',family='serif')
plt.rcParams['xtick.labelsize']=16
plt.rcParams['ytick.labelsize']=16
plt.rcParams['axes.labelsize']=18
plt.rcParams['axes.titlesize']=18
```

# $$ a_{1D} = \frac{\Gamma(-E/2 + 1/4)}{2\Gamma(-E/2 + 3/4)} $$

```python
Einf = -4
Esup = 0.5
DE = 0
fig, ax = plt.subplots(figsize=(8,6))
for _ in range(4):
    E = np.linspace(Einf, Esup, num=1000) + DE
    a1D = gamma(-E/2 + 1/4)/(2*gamma(-E/2+3/4))
    ax.plot(-a1D, E, 'darkblue', lw=2)
    ax.set_xlabel('$-a_{1D}/d_{HO}$')
    ax.set_ylabel('$E[\hbar \omega]$')
    ax.set_title("1D energy degeneracy of two ultracold atoms")
    ax.set_ylim(-4, 10)
    ax.set_xlim(-20, 20)
    DE += 4.5
fig.savefig('1D_energy_degeneracy_2UCAtoms.png', dpi=200)
```
