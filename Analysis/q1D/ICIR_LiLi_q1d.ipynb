# Comparison with Theory

```python
Data = np.loadtxt('Results/ICIR_q1D.txt')
wxwy = Data[0]
asc_020 = Data[1]
asc_200 = Data[2]

Data_theory = np.loadtxt('../../asc_integral/ICIR_Theory.txt')
wxwy_theory = Data_theory[0]
asc_020_theory = Data_theory[1]
asc_200_theory = Data_theory[2]
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(wxwy, asc_020, 'k*', markersize=8)
ax.plot(wxwy, asc_200, 'k*', markersize=8)
ax.fill_between(wxwy, asc_020_theory, asc_200_theory, 'orange', alpha=0.6)
ax.set_xlim(0.997, 1.254)
ax.set_ylim(0.64, 0.83)
ax.set_xlabel('$\omega_x/\omega_y$')
ax.set_ylabel('$a^{ICIR}/d_y$')
plt.grid()
fig.savefig('Results/Figures/ICIR_q1D_Theory_band.png', dpi=200)
```
