## ICIR: $\psi^{(b)}\Phi_{(0, 4, 0)}, \psi^{(b)}\Phi_{(4, 0, 0)}$

```python
wxwy = [1, 1.05, 1.1, 1.15, 1.20, 1.25]
ady_040 = np.array([0.632837560318262, 0.638151740308206 , 0.643270166032872, 0.648634553931570, 0.653803480332060 , 0.656712856778913])
ady_400 = np.array([0.467285274828366, 0.453980577935556, 0.440216756672616, 0.427326030368223, 0.417035196629350, 0.405671307060611])
```

```python
xnew = np.linspace(0, 1.25, num=6)
fit040 = np.polyfit(wxwy, ady_040, deg=1)
fit400 = np.polyfit(wxwy, ady_400, deg=1)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(wxwy, ady_040, 'w*',\
         markersize=16, markeredgecolor='k')
ax.plot(wxwy, ady_400, 'w*',\
         markersize=16, markeredgecolor='k',)
ax.plot(xnew, np.polyval(fit040, xnew), 'r')
ax.plot(xnew, np.polyval(fit400, xnew), 'r')
ax.set_xlim(0.5, 1.254)
#ax.set_ylim(0.50, 0.59)
ax.set_xlabel('$\omega_x/\omega_y$')
ax.set_ylabel('$a_s/d_y$')
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
#plt.grid()
plt.tight_layout()
#fig.savefig('Results/Figures/ICIR_q2dq3d_LiLi_400.png', dpi=200)
```
