```python
__author__ = "@Tssp"
from numpy.polynomial.hermite import hermval
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, num=200)
```

```python
def Hermite_plot(n, x):
    coef = np.zeros(n+1)
    coef[-1] = n
    plt.plot(x, hermval(x, coef))
    plt.ylabel(r'$H_{}(x)$'.format(1), fontsize=16)
    plt.xlabel(r'$x$', fontsize=16)
    plt.xlim(-3, 3)
    plt.ylim(-50, 50)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
leg = []
for i in range(6):
    Hermite_plot(i, x)
    leg.append('n=' + str(i))
    plt.legend(leg)
```

# Rotations

```python
def Hermite_rotations(n, x, theta):
    coef = np.zeros(n+1)
    coef[-1] = n
    y = hermval(x, coef)
    x_new = x * np.cos(theta) - y * np.sin(theta)
    y_new = x * np.sin(theta) + y * np.cos(theta)
    plt.plot(x_new, y_new)
    plt.ylabel(r'$H_{}(x)$'.format(1), fontsize=16)
    plt.xlabel(r'$x$', fontsize=16)
    plt.xlim(-3, 3)
    plt.ylim(-50, 50)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
n=2
Hermite_plot(n, x)
Hermite_rotations(n, x, np.pi)
plt.legend(['original', 'rotated'])
```

# Reflections

```python
fig, ax = plt.subplots(figsize=(8, 6))
n=3
coef = np.zeros(n+1)
coef[-1] = n
Hermite_plot(n, x)
plt.plot(x, hermval(-x, coef))
plt.legend(['original', 'reflected'])
```
